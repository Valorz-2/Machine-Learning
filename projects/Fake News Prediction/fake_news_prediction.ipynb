```python
# Import dependencies

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import re
import string
from nltk.corpus import stopwords # natural language toolkit
from nltk.stem.porter import PorterStemmer # To perform Stemming  
from sklearn.feature_extraction.text import TfidfVectorizer # to convert text into feature vectors
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import os
from dotenv import  load_dotenv

import nltk
nltk.download('stopwords')
```

## Data Preprocessing

```python
#loading the datasets in dataframe
load_dotenv()

train_dataset = os.getenv("news_dataset_train")
print(train_dataset)

news_dataset = pd.read_csv(train_dataset)
print(news_dataset)
```

```python
# replace the null values with empty string 
news_dataset = news_dataset.fillna('')

#merging the author name and news title
news_dataset['content'] = news_dataset['author'] +' '+news_dataset['title']

print(news_dataset['content'])
```

```python
# separating the data and label

X = news_dataset.drop(columns='label',axis=1)
Y = news_dataset['label']
```

Stemming is the process of reducing a word to its root word

eg : actor, actress, acting --> act

```python
port_stem = PorterStemmer()

def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]',' ', content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```

```python
news_dataset['content'] = news_dataset['content'].apply(stemming)
```

```python
# separating the data and label
X = news_dataset['content'].values
Y = news_dataset['label'].values
print(X)
```

```python
print(Y)
Y.shape
```

```python
# converting text into numerical data
vectorizer = TfidfVectorizer() # term frequency inverse document frequency

# counts the no of times a words appears and tells model how important that word is y assigning a number
# sometimes a word that appears several times doesn't have any meaning 

vectorizer.fit(X)
X = vectorizer.transform(X)
```

```python
print(X)
```

```python
# Splitting data into training and testing

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size= 0.2, stratify=Y, random_state=2)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
# training data accuracy
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
print("Accuracy score of training data :", training_data_accuracy)

#testing data accuracy
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, y_test)
print("Accuracy score for test data : ", testing_data_accuracy)
```

Making a predictive system

```python
input_data = X_test[0]

prediction = model.predict(input_data)

print(prediction)

if prediction[0] == 0:
    print("Its a real news")
else:
    print("The news is fake")
```
